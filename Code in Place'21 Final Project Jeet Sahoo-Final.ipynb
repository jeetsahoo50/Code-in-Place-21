### Observations

New York and California have the largest number of purchasers. Only 2 or 3 quantities are bought by most clients. Max 0 to 20 percent discount. The relationship between profit and discount is null. In the west and in the southern area the largest profits and sales are achieved. Sales tables suffer the most loss. In standard class, sales and profit are larger.

### Conclusion

The weak areas that are possible to earn more profit are: the sales of furnishings should be limited and the sales of office supplies and technology increased as it produces very little profit than its sales. We face enormous loss in the sub-categories in tables, so try to decrease the loss. Sales and income in the southern and central regions are reduced. In order to stimulate sales we need provide additional incentives such as discounts and consequently improve profits. We have high value for potential clients in Texas and New York after the biggest sales in California, thus we should concentrate more on these States to make more profit. Few states like North Dakota, Wyoming and West Virginia have less or no sales at all so we also need to focus on these states. The percentage of discount offered should not be more than 25%, otherwise loss is faced.
